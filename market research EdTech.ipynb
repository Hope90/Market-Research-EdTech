**From the above graph, after 2016, there are many missing values.**

**Considering the marketing study should focus on more recent years, therefore, I will focus on the data from 2010 to 2016. The latest available data between 2010 and 2016 will be used.**

```python
value = data_clean[['Country Name','Country Code','Region','Indicator Name','Indicator Code','2010','2011','2012','2013','2014','2015','2016']]
value.head()
```

```python
def last_value(df):
    """
    last_value is a function that aims to return the last valid data of a dataframe

    :param df: dataframe to parse
    :return: last valid data by index
    
    method used: 
    numpy.nan: return a floating point representation of Not a Number.
    pandas.DataFrame.last_valid_index(): return index for last non-NA/null value.
    
    function can be called with another method:
    pandas.DataFrame.apply(): apply a function along an axis of the DataFrame.
    """ 
    if df.last_valid_index() is None:
        return np.nan
    else:
        return df[df.last_valid_index()]

value['Value'] = value.apply(last_value, axis=1)
```

```python
value.head(10)
```

```python
for i in range(len(value)):
  if type(value.at[i,'Value']) == str:
    value.at[i,'Value']=np.NaN
value.head(10)
```

```python
value_clean = value.drop(columns=['2010','2011','2012','2013','2014','2015','2016'])
value_clean.head()
```
